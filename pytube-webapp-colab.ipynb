After you have installed all dependencies, make sure to disable debug mode!  
This step is crucial!
Make sure to change the following lines in [main.py](main.py) from this:

```python
if __name__ == "__main__":
    app.run(debug=True)
```

...to this:  

```python
if __name__ == "__main__":
    app.run()
```
